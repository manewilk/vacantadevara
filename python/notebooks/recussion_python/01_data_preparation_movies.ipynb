```python
# Read sqlite query results into a pandas DataFrame
con = sqlite3.connect("C:/Users/radu_/.bokeh/data/movies.db")
directorsdf = pd.read_sql_query("SELECT userRating, imdbRating, Title, Year, Director FROM omdb, tomatoes WHERE omdb.ID = tomatoes.ID", con)

# Verify that result of SQL query is stored in the dataframe
print(directorsdf.head(20))

con.close()
```

```python
directors = list(directorsdf['Director'].unique())
directors[1:15]
```

```python
directors = ['Alfred Hitchcock', 'Akira Kurosawa', 'Jim Jarmusch', 'Andrei Tarkovsky', 'Larry Cohen', 'Pedro Almodóvar']
```

```python
list(directorsdf['Director'].unique())
```

```python
output_notebook()
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.palettes import Spectral6

p = figure(title = "Movies Ratings and Directors", background_fill_color="#fafafa", 
           tooltips=[("Movie Title", "@Title"), ("Director", "@Director"), ("Year", "@Year")])
p.xaxis.axis_label = 'IMDb Rating Cristinel'
p.yaxis.axis_label = 'Tomato User Rating'

#directors = list(directorsdf['Director'].unique())

for idx, director in enumerate(directors):
    p.circle('imdbRating', 'userRating',
             source=directorsdf[directorsdf['Director']==director], legend_label=director, 
             color=Spectral6[idx], fill_alpha=0.9, size=15, line_color=Spectral6[idx], line_width=2,
            )

p.legend.location = "top_left"
p.legend.click_policy= "hide"
show(p)
```
